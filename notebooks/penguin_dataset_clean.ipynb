```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. CARGA DEL DATASET 
df_original = sns.load_dataset("penguins")
print(f"Dimensiones originales antes de limpiar: {df_original.shape}")

# 2. LIMPIEZA DE DATOS (valor 'NA')
df_clean = df_original.dropna()

# 3. COMPROBACIÓN Y RESUMEN
print(f"Registros finales tras eliminar nulos: {df_clean.shape[0]}")
print(f"Valores nulos restantes en el dataset: {df_clean.isnull().sum().sum()}")

# 4. VISUALIZACIÓN DE LOS DATOS LIMPIOS
display(df_clean.head())
display(df_clean.describe())
```
